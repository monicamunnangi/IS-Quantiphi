```python
allFiles = glob.glob("training_A/p*.psv")

#patients_kalman = pd.DataFrame()
patients = pd.DataFrame()
patient_id = 0

for file_ in allFiles:
    dftmp = pd.read_csv(file_,sep = '|')
    
    patient_data = pd.DataFrame()
    #patient_data_kalman = pd.DataFrame()
    patient_data['pid'] = patient_id
    #patient_data_kalman['pid'] = patient_id
    
    patient_id += 1
    features = dftmp.columns
    
    for feature in features:
        #print(feature)
        if feature == 'SepsisLabel':
            continue
        patient_data[feature] = dftmp[feature]
        '''
        try :
            ro.globalenv['x'] = dftmp[feature]
            y = ro.r("na.kalman(x, model =\"auto.arima\", smooth = TRUE)")
            patient_data_kalman[feature] = y
        except :
            patient_data_kalman[feature] = dftmp[feature]
        '''     
    #print(patient_data['pid'])
    patient_data['pid'] = patient_id
    patient_data.fillna(method = 'backfill', inplace=True)
    patient_data.fillna(method = 'ffill', inplace=True)
    patient_data['SepsisLabel'] = dftmp['SepsisLabel']
    patients = pd.concat([patients, patient_data])
    
    #patient_data_kalman['pid'] = patient_id
    #patient_data_kalman.fillna(method = 'backfill', inplace=True)
    #patient_data_kalman.fillna(method = 'ffill', inplace=True)
    #patient_data_kalman['SepsisLabel'] = dftmp['SepsisLabel']
    #patients_kalman = pd.concat([patients_kalman, patient_data_kalman])
```
